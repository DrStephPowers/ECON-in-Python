#Graphing Supply and Demand

```python
P = (50, 45, 40, 35, 30, 25, 20, 15, 10)
```

```python
Q_S = (180, 173, 169, 150, 143, 130, 125, 110, 104)
```

```python
Q_D = (124, 137, 143, 154, 164, 168, 174, 185, 193)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Q_S, P, marker='o')
plt.plot(Q_D, P, marker='o')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.legend(["Supply", "Demand"], bbox_to_anchor=(0.75, -0.15), ncol=2)
plt.grid()
plt.show()
```

P = -40 + 0.49 Qs

###Rearrange and Solve

40 = 0.49 Q - P

126 = 0.6Q + P

```python
A = np.array([[.49, -1], [0.6, 1]])
B = np.array([40, 126])
```

```python
np.linalg.solve(A, B)
```

Solution: Q = 152, P = 34.6

## Plot fitted lines

```python
Qd = np.linspace(100, 200, 10)
Pd =  126 - 0.60*Qd

Qs = np.linspace(100, 200, 10)
Ps = -40 + 0.49*Qs
```

```python
plt.plot(Qs, Ps, marker='o')
plt.plot(Qd, Pd, marker='o')
plt.axvline(x=152, color="black")
plt.hlines(y=34.6, xmin=100, xmax=200, color="black")
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.legend(["Supply", "Demand"], bbox_to_anchor=(0.75, -0.15), ncol=2)
plt.grid()
plt.show()
```
