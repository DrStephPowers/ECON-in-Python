```python
import matplotlib.pyplot as plt
```

```python
Can14 = (0.072, 0.131, 0.177, 0.235, 0.385)
```

```python
US14 = (0.031, 0.082, 0.143, 0.232, 0.512)
```

```python
Per = (0, 0.20, 0.40, 0.60, 0.80, 1.0)
```

```python
Imper = (0, 0, 0, 0, 0, 0)
```

```python
Quin = (0, 0.20, 0.40, 0.60, 0.80, 1.0)
```

```python
plt.plot(Quin, Per, marker='o')
plt.plot(Quin, Imper, marker='o')
plt.xlabel('Population')
plt.ylabel('Cumulative Income')
plt.legend(["Perfect Equality", "Perfect Inequality"], bbox_to_anchor=(0.75, -0.15), ncol=2)
plt.grid()
plt.show()
```

```python
cumCan = []
init = 0
```

```python
for i in Can14:
  init += i
  cumCan.append(init)
```

```python
cumCan
```

```python
cumCan.insert(0, 0)
```

```python
plt.plot(Quin, Per, marker='o')
plt.plot(Quin, Imper, marker='o')
plt.plot(Quin, cumCan, marker='o')
plt.xlabel('Population')
plt.ylabel('Cumulative Income')
plt.legend(["Perfect Equality", "Perfect Inequality", "Canada"], bbox_to_anchor=(0.75, -0.15), ncol=3)
plt.grid()
plt.show()
```
