## AD and AS


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
CPI = (96, 97, 98, 99, 100, 101, 102, 103, 104, 105)
```

```python
AD = (1145, 1125, 1105, 1085, 1065, 1045, 1025, 1005, 985, 965)
```

```python
AS = (880, 940, 965, 985, 1000, 1015, 1025, 1033, 1040, 1045)
```

```python
plt.plot(AD, CPI, marker='o')
plt.plot(AS, CPI, marker='o')
plt.xlabel('rGDP')
plt.ylabel('CPI')
plt.grid()
plt.show()
```

```python
Pot = np.repeat(1025, len(CPI))
```

```python
plt.plot(AD, CPI, marker='o')
plt.plot(AS, CPI, marker='o')
plt.plot(Pot, CPI, linestyle='dashed')
plt.xlabel('rGDP')
plt.ylabel('CPI')
plt.grid()
plt.show()
```
