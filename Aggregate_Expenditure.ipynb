## Spending Multiplier

```python
C = lambda y: 5 + 0.6*y
I = lambda y: 75
G = lambda y: 160
NX = lambda y: 60 - 0.1*y
```

```python
AE = lambda y: C(y) + I(y) + G(y) + NX(y)
```

```python
AE(400)
```

```python
NI = (0, 100, 200, 300, 400, 500, 600, 700, 800)
```

```python
NI = np.array(NI)
```

```python
x = []
```

```python
for i in NI:
   x1 = AE(i)
   x.append(x1)
```

```python
x
```

```python
plt.plot(NI, x, marker='o')
plt.plot(NI, NI, marker='o')
plt.xlabel('National Income')
plt.ylabel('Aggregate Expenditures')
plt.grid()
plt.show()
```

```python
NI = np.array(NI).reshape((-1, 1))
x = np.array(x)
```

```python
model = LinearRegression().fit(NI, x)
```

```python
model.coef_
```

```python
model.intercept_
```

AE = 300 + 0.5 Y
